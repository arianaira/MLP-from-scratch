```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean = 0  
std_dev = 1
size = 1000

x_values = np.random.normal(mean, std_dev, size)
```

```python
sns.distplot(x_values, kde=False, rug=True, color="green")
plt.title('Distribution Plot for x values')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def identity(x):
    return x
```

```python
def dev_identity(x):
    return np.ones_like(x)
```

```python
y_identity = identity(x_values)
y_dev_identity = dev_identity(x_values)
sns.set(style="whitegrid") 
sns.lineplot(x=x_values, y=y_identity, label='Identity Function')
sns.lineplot(x=x_values, y=y_dev_identity, label='Derivative of Identity Function')
plt.title('Identity Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.show()
```

```python
sns.set(style="whitegrid") 
y_identity = identity(x_values)
sns.distplot(y_identity, kde=False, rug=True)
plt.title('Distribution Plot for Identity Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_identity = dev_identity(x_values)
sns.distplot(y_dev_identity, kde=False, rug=True)
plt.title('Distribution Plot for dev Identity Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def dev_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

y_sigmoid = sigmoid(x_values)
y_dev_sigmoid = dev_sigmoid(x_values) 

sns.lineplot(x=x_values, y=y_sigmoid, label='Sigmoid Function')
sns.lineplot(x=x_values, y=y_dev_sigmoid, label='Derivative of Sigmoid Function')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
y_sigmoid = sigmoid(x_values)
sns.distplot(y_sigmoid, kde=False, rug=True)
plt.title('Distribution Plot for sigmoid Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_sigmoid = dev_sigmoid(x_values)
sns.distplot(y_dev_sigmoid, kde=False, rug=True)
plt.title('Distribution Plot for dev sigmoid Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

def dev_tanh(x):
    return 1 - np.tanh(x) ** 2

y_tanh = tanh(x_values) 
y_dev_tanh = dev_tanh(x_values) 

sns.lineplot(x=x_values, y=y_tanh, label='tanh Function')
sns.lineplot(x=x_values, y=y_dev_tanh, label='Derivative of tanh Function')

plt.title('tanh Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
y_Tanh = tanh(x_values)
sns.distplot(y_Tanh, kde=False, rug=True)
plt.title('Distribution Plot for tanh Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_tanh = dev_tanh(x_values)
sns.distplot(y_dev_tanh, kde=False, rug=True)
plt.title('Distribution Plot for dev tanh Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

def dev_relu(x):
    return np.where(x > 0, 1, 0)

y_relu = relu(x_values)
y_dev_relu = dev_relu(x_values)

sns.lineplot(x=x_values, y=y_relu, label='ReLU Function')
sns.lineplot(x=x_values, y=y_dev_relu, label='Derivative of ReLU Function')
plt.title('ReLU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
y_relu = relu(x_values)
sns.distplot(y_relu, kde=False, rug=True)
plt.title('Distribution Plot for relu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_relu = dev_relu(x_values)
sns.distplot(y_dev_relu, kde=False, rug=True)
plt.title('Distribution Plot for dev relu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)


def dev_leaky_relu(x, alpha=0.1):
    return np.where(x > 0, 1, alpha)

y_leaky_relu = leaky_relu(x_values)
y_dev_leaky_relu = dev_leaky_relu(x_values) 
sns.lineplot(x=x_values, y=y_leaky_relu, label='Leaky ReLU Function')
sns.lineplot(x=x_values, y=y_dev_leaky_relu, label='Derivative of Leaky ReLU Function')

plt.title('Leaky ReLU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.show()
```

```python
y_leaky_relu = leaky_relu(x_values)
sns.distplot(y_leaky_relu, kde=False, rug=True)
plt.title('Distribution Plot for leaky relu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_leaky_relu = dev_leaky_relu(x_values)
sns.distplot(y_dev_leaky_relu, kde=False, rug=True)
plt.title('Distribution Plot for dev leaky relu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def step(x):
    return np.where(x >= 0, 1, 0)

def dev_step(x):
    return np.zeros_like(x)

y_step = step(x_values) 
y_dev_step = dev_step(x_values) 

sns.lineplot(x=x_values, y=y_step, label='Step Function')
sns.lineplot(x=x_values, y=y_dev_step, label='Derivative of Step Function')

plt.title('Step Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend() 
plt.show()
```

```python
y_step = step(x_values)
sns.distplot(y_step, kde=False, rug=True)
plt.title('Distribution Plot for step Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_step = dev_step(x_values)
sns.distplot(y_dev_step, kde=False, rug=True)
plt.title('Distribution Plot for dev step Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def softplus(x):
    return np.log(1 + np.exp(x))

def dev_softplus(x):
    return 1 / (1 + np.exp(-x))

y_softplus = softplus(x_values)
y_dev_softplus = dev_softplus(x_values) 

sns.lineplot(x=x_values, y=y_softplus, label='Softplus Function')
sns.lineplot(x=x_values, y=y_dev_softplus, label='Derivative of Softplus Function')

plt.title('Softplus Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()  
plt.show()
```

```python
y_dev_softplus = dev_softplus(x_values)
sns.distplot(y_dev_softplus, kde=False, rug=True)
plt.title('Distribution Plot for dev softplus Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
def hard_tanh(x):
    return np.maximum(-1, np.minimum(1, x))

def dev_hard_tanh(x):
    return np.where(np.logical_and(x >= -1, x <= 1), 1, 0)

y_hard_tanh = hard_tanh(x_values) 
y_dev_hard_tanh = dev_hard_tanh(x_values) 

sns.lineplot(x=x_values, y=y_hard_tanh, label='Hard tanh Function')
sns.lineplot(x=x_values, y=y_dev_hard_tanh, label='Derivative of Hard tanh Function')

plt.title('Hard tanh Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
y_hard_tanh = hard_tanh(x_values)
sns.distplot(y_hard_tanh, kde=False, rug=True)
plt.title('Distribution Plot for hard_tanh Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_hard_tanh = dev_hard_tanh(x_values)
sns.distplot(y_dev_hard_tanh, kde=False, rug=True)
plt.title('Distribution Plot for dev hard_tanh Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.special import erf

def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))

def dev_gelu(x):
    return 0.5 * (1 + erf(x / np.sqrt(2))) + 0.5 * x * np.exp(-0.5 * (x / np.sqrt(2))**2) / np.sqrt(2*np.pi)


y_gelu = gelu(x_values) 
y_dev_gelu = dev_gelu(x_values)  

sns.lineplot(x=x_values, y=y_gelu, label='GELU Function')
sns.lineplot(x=x_values, y=y_dev_gelu, label='Derivative of GELU Function')

plt.title('GELU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
y_gelu = gelu(x_values)
sns.distplot(y_gelu, kde=False, rug=True)
plt.title('Distribution Plot for gelu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
y_dev_gelu = dev_gelu(x_values)
sns.distplot(y_dev_gelu, kde=False, rug=True)
plt.title('Distribution Plot for dev gelu Function')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```
